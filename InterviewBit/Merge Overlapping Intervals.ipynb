```python
# Definition for an interval.
# class Interval:
#     def __init__(self, s=0, e=0):
#         self.start = s
#         self.end = e

class Solution:
    # @param intervals, a list of Intervals
    # @return a list of Interval                  #### main #### Merge Overlapping Intervals
    def merge(self, A):
        A = [[obj.start,obj.end] for obj in A]
        A.sort(key=lambda x:x[0])
        ans = [A[0]]
        for i,j in A[1:]:
            if ans[-1][1]>=i:
                ans[-1][1]=max(ans[-1][1],j)
            else:
                ans+=[[i,j]]
        #print(ans)
        return [Interval(i,j) for i,j in ans]
        
```
